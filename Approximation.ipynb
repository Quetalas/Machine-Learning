# Цель: Аппроксимировать функцию

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.linalg
```

```python
def f(x):
    return np.sin(x/5.0)*np.exp(x/10.0) + 5.0*np.exp(-x/2.0)
```

```python
f(np.array([0,1,2,3]))
```

```python
plt.plot(np.linspace(-20,21,1000), f(np.linspace(-20,21,1000)))
plt.show()
```

```python
#  матрица положенияй по оси Х и ответов Y
X = np.array([[1,1],[1,15]])
Y = f(np.array([1,15]))
```

```python
print(X)
print(Y)
```

```python
W = scipy.linalg.solve(X, Y)
```

```python
g = lambda X: W[0] + X.dot(W[1:])
```

```python
plt.plot(np.linspace(1,15,1000), f(np.linspace(1,15,1000)))
plt.plot(np.linspace(1,15,1000), g(np.linspace(1,15,1000).reshape(1000,1)), c ='r')
plt.show()
```

```python
#  матрица положенияй по оси Х и ответов Y
X = np.array([[1,1,1], [1,8,64], [1,15,225]])
Y = f(np.array([1, 8, 15]))
print(X)
print(Y)
W = scipy.linalg.solve(X, Y)
print(W)
```

```python
g = lambda X: W[0] + X.dot(W[1:])
```

```python
X = []
for x in np.linspace(1,15,1000):
    X.append([x, x**2])
X = np.array(X)
plt.plot(np.linspace(1,15,1000), f(np.linspace(1,15,1000)))
plt.plot(np.linspace(1,15,1000), g(X), c ='r')
plt.show()
```

```python
#  матрица положенияй по оси Х и ответов Y
X = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
Y = f(np.array([1, 4, 10, 15]))
print(X)
print(Y)
W = scipy.linalg.solve(X, Y)
print(W)
```

```python
X = []
for x in np.linspace(1,15,1000):
    X.append([x, x**2, x**3])
X = np.array(X)
plt.plot(np.linspace(1,15,1000), f(np.linspace(1,15,1000)))
plt.plot(np.linspace(1,15,1000), g(X), c ='r')
plt.show()
```
