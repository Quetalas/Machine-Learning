```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from rosenblatt import Perceptron
```

```python
from mnist import MNIST
mndata = MNIST(r'C:\Users\Zhenia\Desktop\mnist_data')
images, labels = mndata.load_training()
test_images, test_labels = mndata.load_testing()
```

```python
ppn = Perceptron()
```

```python
ppn.fit(np.array(images[:1000]), np.array(labels[:1000]))
```

```python
ppn.predict(np.array(images[0]))
```

```python
ppn.accuracy(np.array(images[:1000]), np.array(labels[:1000]))
```
